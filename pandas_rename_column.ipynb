## Pandas: Renaming Column and Indices

```python
import pandas as pd
df_exams = pd.read_csv("datasets/StudentsPerformance.csv")
df_exams.head()
```

```python
# Renaming one or more columns

df_exams.rename(columns={"gender": "sex", 
                         "race/ethnicity": "ethnicity", 
                         "test preparation course": "test prep", 
                         "math score": "math", 
                         "reading score": "reading", 
                         "writing score": "writing"}, inplace=True)
df_exams.head()
```

```python
# Renaming indices (0, 1, 2...)

df_exams.rename(index={0: "A", 1: "B", 2: "C"}, inplace=True)
df_exams.head()
```
