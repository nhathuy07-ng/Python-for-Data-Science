## Maths Operations in Pandas

```python
import pandas as pd
```

```python
df_exams = pd.read_csv("datasets/StudentsPerformance.csv")
df_exams.head()
```

### **Operations in Columns**

```python
# Calculate total sum of a column
df_exams['math score'].sum()
```

```python
# Calculate count, mean, standard deviation, max and min
df_exams['math score'].count(), df_exams['math score'].mean(), df_exams['math score'].std(), df_exams['math score'].max(), df_exams['math score'].min() 
```

```python
# Calculate all stats of a column with .describe()
df_exams['math score'].describe()
```

```python
df_exams.describe()
```

### **Operations in Rows**

```python
# Calculating the sum in a row

# Example: sum of math score + reading score + writing score for each row.
df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']
```

```python
# Calculating average score

exam_average = (df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']) / 3
```

```python
# Add average rounded score to DataFrame and show DataFrame

df_exams['average'] = exam_average.round(2)
df_exams
```
