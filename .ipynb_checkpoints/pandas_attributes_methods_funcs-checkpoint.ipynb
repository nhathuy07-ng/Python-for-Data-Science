## Pandas: Attributes, Methods and Functions

```python
import pandas as pd
import numpy as np
```

```python
df_exams = pd.read_csv('datasets/StudentsPerformance.csv')
df_exams.head()
```

### Basic Attributes of a DataFrame

```python
# Shape: (rows, cols)
df_exams.shape
```

```python
# Index attribute
# RangeIndex(start, stop (last element), step)
df_exams.index
```

```python
# Column attribute: List of name of columns
df_exams.columns
```

```python
# Data types of each column
# strings --> type object
df_exams.dtypes
```

### Methods

```python
# show first 5 rows
df_exams.head(5)
```

```python
# showing the info of the dataframe
df_exams.info()
```

```python
# Describe basic statistics of the dataframe
# Count, Mean, Standard Deviation, Min/Max, Quartiles
df_exams.describe()
```

### Functions

```python
# Length (number of rows) of the dataframe
len(df_exams)
```

```python
# Highest index of the dataframe: corresponds to the last entry on the dataframe
max(df_exams.index)
```

```python
# Lowest index for the dataframe
min(df_exams.index)
```

```python
# Obtain the data type of the DataFrame
type(df_exams)
```

```python
# Rounding values of the dataset
round(df_exams, 2)
```

## **Selecting columns**
### Select one column

```python
# Select column with []
df_exams['gender']
```

```python
# A column has a type of Series
type(df_exams['gender'])
```

```python
# Select column with dot notation
df_exams.gender
```

### Select two or more columns

```python
  # Using the [[]] (nested square brackets)
df_exams[["test preparation course", "math score", "reading score", "writing score"]]
```

```python
type(df_exams[["math score", "reading score"]])
```

```python
# Nested square brackets would always return a DataFrame
type(df_exams[["test preparation course"]])
```

## **Adding a new column**

```python
# Add a new column with a scalar value.
# Every row of the new column shares the same value
df_exams['language score'] = 85
df_exams
```

```python
# Add new column with an array
# The dataframe has 1000 rows, hence 1000 elements is needed

# 1000 random elements in a list in range [50, 85] ("high" argument exclusive)
lang_score = np.random.randint(50, 85+1, 1000) # length must either be a scalar or match index's length
df_exams['language score'] = lang_score
df_exams
```
