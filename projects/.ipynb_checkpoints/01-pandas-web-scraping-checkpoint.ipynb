```python
# Reading 1 csv file from the site

df = pd.read_csv("https://football-data.co.uk/new/CHN.csv")
df.head()
```

```python
# Renaming columns
df.rename(columns={"Season": "Year", "HG": "Home Goal", "AG": "Away Goal"}, inplace=True)
df.head(3)
```

```python
df.sort_values("Year", ascending=False, inplace = True)
df[["Date", "Time", "Home", "Away", "Home Goal", "Away Goal", "Res"]]
```

```python
# Reading multiple CSVs from Pandas

# Step 1: Get all the links of the France Football result from 2020-2025
root = "https://football-data.co.uk/mmz4281"
urls = []
for i in range(25, 19, -1):
    url = root + "/" + str(i) + str(i+1) + "/" + "F1.csv"
    urls.append(url)
    url = root + "/" + str(i) + str(i+1) + "/" + "F2.csv"
    urls.append(url)
```

```python
# Read multiple CSVs and append

dframes = []

for url in urls:
    dframes.append(
        pd.read_csv(url)
    )
    
```
