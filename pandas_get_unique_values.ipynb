## Pandas: Get and count unique values.

```python
import pandas as pd

df = pd.read_csv("datasets/laptop_price.csv", encoding="unicode-escape")
```

#### `unique()`: Return unique values of Series object in order of appearance.

```python
df.Company.unique()
```

```python
df.Inches.unique()
```

```python
# Get length of unique elements
len(df.Inches.unique())
```

#### `nunique()`: Return number of unique elements in a series (excluding N/A)

```python
df.Inches.nunique() # Same as len(df.Inches.unique())
```
