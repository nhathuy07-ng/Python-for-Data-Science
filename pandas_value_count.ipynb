## `value_counts()` in Pandas

```python
import pandas as pd

df_exams = pd.read_csv("datasets/StudentsPerformance.csv")
```

```python
# Counting elements of a row

# len function
len(df_exams['gender'])

# .count() method
df_exams['gender'].count()
```

```python
# counting occurences for each value

# Example: counting number of male and female students
df_exams['test preparation course'].value_counts()
```

```python
# get the relative frequency of each category: divide all counts by sum of counts
df_exams['test preparation course'].value_counts(normalize=True)
```

```python
# round relative frequency to 2 decimals
df_exams['parental level of education'].value_counts(normalize=True).round(2)
```
