## **Adding a new column**

### Using ["column"] notation

```python
# Add a new column with a scalar value.
# Every row of the new column shares the same value
df_exams['language score'] = 85
df_exams
```

```python
# Add new column with an array
# The dataframe has 1000 rows, hence 1000 elements is needed

# 1000 random elements in a list in range [0, 100] ("high" argument exclusive)
lang_score = np.random.randint(0, 100+1, 1000) # length must either be a scalar or match index's length
df_exams['language score'] = lang_score
df_exams
```

```python
# Still like above, using random floats from 0 to 100
lang_score = np.random.uniform(0, 100+1, 1000) # 'high' value exclusive
df_exams['language score'] = lang_score
df_exams
```

### Using assign()

- Add multiple columns in a single line
- For overwriting values of existing columns (best practice)

Returns a new object of the `DataFrame` with all the **original columns + the new ones**.

```python
# Create random numbers for our new columns (1000 rows)
score1 = np.random.randint(1, 100, size=1000)
score2 = np.random.randint(1, 100, size=1000)

# Create series
series1 = pd.Series(score1)
series2 = pd.Series(score2)

# assign() to add multiple columns. This create a new copy for the DataFrame
df_exams.assign(score1=series1, score2=series2)
```

```python
# The original DataFrame is NOT modified
df_exams
```

### Using insert()
- Insert a new column at a specific position or index.
- Updates DataFrame in-place.

```python
# insert(position, column name, Series) updates DataFrame in-place
# Fails if column name is duplicate
try:
    df_exams.insert(1, 'test_misc', series1)
except:
    pass
df_exams[["gender", "test_misc"]]
```
