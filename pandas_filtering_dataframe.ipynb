### **Filtering DataFrame based on multiple conditions**

- `&` operator: AND operator
- `|` operator: OR operator

```python
# AND: Find Apple laptops that cost more than 2000 EUR

condition1 = df.Company == "Apple"
condition2 = df.Price_euros > 2000
df[condition1 & condition2]
```

```python
# OR: Find laptops that are EITHER from Apple OR Dell

condition1 = df.Company == "Apple"
condition2 = df.Company == "Dell"
df[condition1 | condition2]
```

```python
# Verify
df[condition1 | condition2].value_counts("Company")
```

```python
# Exercise: find laptops from Apple or Dell that cost over 2000 Euros

cond1 = (df["Company"] == 'Apple') | (df["Company"] == 'Dell')
cond2 = df["Price_euros"] > 2000

edf = df[cond1 & cond2]
edf
```

```python
# Verify
edf.value_counts("Company"), edf.value_counts("Price_euros")
```

### **Filter with lists using `isin()`**

#### Single filtering

```python
df2 = pd.read_csv("datasets/laptop_price.csv", encoding="unicode-escape")
```

```python
# Single filtering

# Select Apple or HP laptops
df2.Company.isin(["Apple", "HP"])
```

```python
# Filter DF
df2[df2.Company.isin(["Apple", "HP"])]
```

```python
# Verify
df2[df2.Company.isin(["Apple", "HP"])].value_counts("Company")
```

#### Multiple Filtering

```python
# Find notebooks or ultrabooks from Apple or HP

cond1 = df2.Company.isin(["Apple", "HP"])
cond2 = df2.TypeName.isin(["Notebook", "Ultrabook"])

cond1 & cond2
```

```python
df2[cond1 & cond2]
```

```python
df2[cond1 & cond2].value_counts(["Company", "TypeName"])
```
